```python
import pandas as pd
import numpy as np
```

```python
DS = pd.read_excel (r'C:\Users\ARIS\Documents\uji-simak-2101.xls', sheet_name=input("sheet name simak:"), header=None)
```

```python
DP = pd.read_excel (r'C:\Users\ARIS\Documents\uji-simak-2101.xls', sheet_name=input("sheet name prima:"), header=None)
```

```python
def sumR(D,r = 1):
    return D[[r]].sum()

def komparasiDenganDesc(D1,D2,r=0):
    return pd.merge(D1,D2,on=r,how='outer',indicator=True)
```

```python
DSG = DS.groupby(by=0).sum()
DSG
```

```python
DResCompare = komparasiDenganDesc(DSG,DP)
DResCompare.loc[:,'selisih'] = DResCompare['1_y'].fillna(0) - DResCompare['1_x'].fillna(0)
DResCompare.loc[:,'kodeTrans'] = np.where(DResCompare['selisih'] < 0, '|P02|', '|P01|')
DResCompare.loc[:,'abs_selisih'] = DResCompare['selisih'].abs()
DResCompare.loc[:,'nama item'] = "|" + DResCompare[4] + "|"
DResCompare.loc[:,'satuan'] = "|" + DResCompare[5] + "|"
DResCompare.loc[:,'hna'] = DResCompare[6]

idTrans = '|024040100415448241KD2021Y0001'
idx = 0
#DResCompare.loc[:,'id_trans'] = idTrans + (np.where(DResCompare['selisih'] < 0,'K','M')) + '|'

DResCompare['id_trans'] = DResCompare.iloc[:,1:].astype(str).\
apply(lambda x: idTrans ,axis=1)
DResCompare['mk_trans'] = DResCompare.iloc[:,1:].astype(str).\
apply(lambda x: ('K' if x['kodeTrans'] == '|P02|' else 'M') + '|' ,axis=1)
DResCompare['tgl_trans'] = DResCompare.iloc[:,1:].astype(str).\
apply(lambda x: ('31-12-2021 23:00:00' if x['kodeTrans'] == '|P02|' else '01-12-2021 23:00:00'),axis=1)

DResCompare.rename(columns={"key_0": "kode gabung", "1_x": "qty persediaan", "1_y": "qty prima"\
                           }, inplace = True)

DResCompare[['kode gabung','qty persediaan','qty prima',2,3,'nama item','satuan','hna',\
             '_merge','selisih','kodeTrans','abs_selisih','id_trans','mk_trans','tgl_trans']]\
[((DResCompare['_merge'] != 'both') | ((DResCompare['_merge'] == 'both')\
            & (DResCompare['qty persediaan'] != DResCompare['qty prima']))) & (DResCompare['selisih'] != 0)]\
    #.to_excel (r'Documents\hasil-uji-simak-'+input('periode :')+'.xls', index = False, header=True)
```
